```python
##### ======================================================
##### CSI 외부의뢰시험 데이터가 들어왔을때, 건설공사 품질관리 업무지침 별표2. 시험기준의 공종으로 분류하는 모델
##### - 최초 CSI의 3개 칼럼(시료명/종목/방법)을 법규 테이블의 종별, 시험종목(상세), 시험방법 칼럼과 최근접 매칭(TF-IDF)을 앙상블해 만든 공종을 분류
##### - CSI(df1)의 모든 칼럼(범주코드화) + 유사도 점수(3개) + 평균(1개) 모두를 피처로 사용하여 RF 모델로 학습. 
##### - CSI 외부의뢰시험 데이터 (행) 단위로 "공종"을 예측. Ground truth Label은 TF-IDF 매칭으로 구성하고, 모든 칼럼 고려 RF 모델을 통해 예측 정확도를 높임.
##### - 정확도 99% 이상으로 매우 높음
##### ======================================================

# ==== 설정 ====
SEED     = 42
TEST_SIZE= 0.2
ENCODING = "utf-8"   # cp949/euc-kr 환경이면 변경
PATH_DF1 = "csi_preprocessing_output.csv"
PATH_DF2 = "parsed_tables_all_pages_with_ordered_triggers.csv"

# ==== import ====
import os, numpy as np, pandas as pd, matplotlib as mpl, matplotlib.pyplot as plt
from IPython.display import display
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
from sklearn.ensemble import RandomForestClassifier

# ==== 한글 폰트(NanumGothic 고정) ====
def set_korean_font():
    plt.rcParams["font.family"] = "NanumGothic"
    plt.rcParams["axes.unicode_minus"] = False
set_korean_font()

# ==== 함수정의 ====
def tx(s: pd.Series) -> pd.Series:
    """문자열 변환 + 공백 트리밍 + 결측 치환"""
    return s.fillna("").astype(str).map(str.strip)

def join_cols(df: pd.DataFrame, cols, sep=" "):
    """여러 칼럼을 공백으로 합쳐 시험종목_join 생성"""
    cols = [c for c in cols if c in df.columns]
    if not cols: return pd.Series([""]*len(df), index=df.index)
    return df[cols].astype(str).fillna("").apply(
        lambda r: sep.join([v for v in r if v and v.lower()!="nan"]), axis=1
    )

def tfidf_cos(left: pd.Series, right: pd.Series, analyzer="char_wb", ngram_range=(2,5)) -> np.ndarray:
    """TF-IDF(char n-gram) + 코사인 유사도 행렬 [len(left) x len(right)]"""
    L, R = tx(left), tx(right)
    vec = TfidfVectorizer(analyzer=analyzer, ngram_range=ngram_range, min_df=1, max_df=1.0)
    vec.fit(pd.concat([L, R], axis=0))
    return cosine_similarity(vec.transform(L), vec.transform(R))

def ensemble3(a_lab, b_lab, c_lab, a_s, b_s, c_s):
    """3채널 다수결 → 동률이면 해당 라벨의 채널 점수 최대값으로 결정"""
    labs = [a_lab, b_lab, c_lab]
    uniq, cnts = np.unique(labs, return_counts=True)
    winners = uniq[cnts == cnts.max()]
    if len(winners) == 1:
        return winners[0]
    best_lab, best_sc = None, -1.0
    for w in winners:
        sc = max([s for lab, s in [(a_lab,a_s),(b_lab,b_s),(c_lab,c_s)] if lab==w])
        if sc > best_sc:
            best_lab, best_sc = w, sc
    return best_lab

# ==== 1) 데이터 로드 ====
print("1) 파일 로드")
df1 = pd.read_csv(PATH_DF1, encoding=ENCODING, dtype=str, low_memory=False)
df2 = pd.read_csv(PATH_DF2, encoding=ENCODING, dtype=str, low_memory=False)
print(f"- DF1: {df1.shape}  |  DF2: {df2.shape}")

# df2: 시험종목_join 생성 + 안전 보정
df2["시험종목_join"] = join_cols(df2, ["시험종목_1","시험종목_2","시험종목_3"])
for c in ["종별","시험방법","시험종목_join","공종"]:
    if c not in df2.columns: df2[c] = ""


# ==== 2) 세 채널 유사도 → Top-1 후보 라벨 ====
print("\n2) 유사도 계산 & 채널별 후보 라벨 추출")
# (a) 시료명(생산국) ↔ 종별
sim_a = tfidf_cos(df1.get("시료명(생산국)", pd.Series([""]*len(df1))), df2["종별"])
idx_a = sim_a.argmax(axis=1)
scr_a = sim_a[np.arange(sim_a.shape[0]), idx_a]
cand_a = df2.loc[idx_a, "공종"].to_numpy()

# (b) 종목 ↔ 시험종목_join
sim_b = tfidf_cos(df1.get("종목", pd.Series([""]*len(df1))), df2["시험종목_join"])
idx_b = sim_b.argmax(axis=1)
scr_b = sim_b[np.arange(sim_b.shape[0]), idx_b]
cand_b = df2.loc[idx_b, "공종"].to_numpy()

# (c) 방법 ↔ 시험방법
sim_c = tfidf_cos(df1.get("방법", pd.Series([""]*len(df1))), df2["시험방법"])
idx_c = sim_c.argmax(axis=1)
scr_c = sim_c[np.arange(sim_c.shape[0]), idx_c]
cand_c = df2.loc[idx_c, "공종"].to_numpy()


# ==== 3) 앙상블 라벨 y ====
print("\n3) 3채널 앙상블 라벨 생성")
y = np.array([
    ensemble3(cand_a[i], cand_b[i], cand_c[i], scr_a[i], scr_b[i], scr_c[i])
    for i in range(len(df1))
])


# ==== 4) 피처: df1 모든 칼럼(범주화) + 유사도 4개 ====
print("\n4) 피처 구성 (df1 모든 칼럼 + 유사도 4개)")
X_df = pd.DataFrame(index=df1.index)
for col in df1.columns:
    X_df[f"col::{col}"] = pd.Categorical(tx(df1[col])).codes
X_df["sim_시료명-종별"]   = scr_a
X_df["sim_종목-시험종목"] = scr_b
X_df["sim_방법-시험방법"] = scr_c
X_df["sim_평균"]         = (scr_a + scr_b + scr_c) / 3.0
feature_names = list(X_df.columns)
print(f"- 사용 Feature 수: {len(feature_names)}")


# ==== 5) 학습과 테스트 셋 80:20 분할 ====
print("\n5) Train/Test 80:20 분할")
stratify = y if len(np.unique(y)) > 1 else None
X_train, X_test, y_train, y_test = train_test_split(
    X_df.values, y, test_size=TEST_SIZE, random_state=SEED, stratify=stratify
)
idx_train, idx_test = train_test_split(
    df1.index, test_size=TEST_SIZE, random_state=SEED, stratify=stratify
)
print(f"- Train: {X_train.shape}, Test: {X_test.shape}, Classes: {len(np.unique(y))}")


# ==== 6) RandomForest 학습 & 예측 ====
print("\n6) RandomForest 학습")
rf = RandomForestClassifier(
    n_estimators=600, max_depth=None,
    min_samples_split=2, min_samples_leaf=1,
    max_features="sqrt",
    random_state=SEED, n_jobs=-1
)
rf.fit(X_train, y_train)
y_pred = rf.predict(X_test)
try:
    proba = rf.predict_proba(X_test)
    cls_to_idx = {c:i for i,c in enumerate(rf.classes_)}
    y_conf = np.array([proba[i, cls_to_idx[y_pred[i]]] for i in range(len(y_pred))])
except Exception:
    y_conf = np.full(len(y_pred), np.nan)


# ==== 7) 평가 ====
print("\n7) 평가 (Test)")
acc  = accuracy_score(y_test, y_pred)
f1m  = f1_score(y_test, y_pred, average="macro")
print(f"- Accuracy: {acc:.4f}")
print(f"- Macro F1: {f1m:.4f}")
print("\n[Classification Report]")
print(classification_report(y_test, y_pred))


# ==== 8) 시각화: (1) 클래스별 정확도 (2) 피처 중요도 Top-20 ====
print("\n8) 시각화")
# (1) 클래스별 정확도
labels = sorted(np.unique(np.concatenate([y_train, y_test])))
cm = confusion_matrix(y_test, y_pred, labels=labels)
per_class_acc = (np.diag(cm) / cm.sum(axis=1).clip(min=1)).astype(float)

plt.figure(figsize=(8, max(3.5, len(labels)*0.35)))
plt.barh(labels, per_class_acc)
plt.title("클래스별 정확도 (Test)")
plt.xlabel("Accuracy")
plt.tight_layout()
plt.show()

# (2) 피처 중요도 Top-20
imp = pd.Series(rf.feature_importances_, index=feature_names).sort_values(ascending=False)
topN = min(20, len(imp))
plt.figure(figsize=(8, max(3.5, topN*0.35)))
imp.head(topN).sort_values().plot(kind="barh")
plt.title("RandomForest Feature Importance (Top-20)")
plt.xlabel("Importance")
plt.tight_layout()
plt.show()

# ==== 9) 테스트 예측 결과(원본 칼럼 포함) ====
print("\n9) 테스트 예측 결과 미리보기 (원본 칼럼 포함)")
test_pred = df1.loc[idx_test].copy()
test_pred["true_공종"] = y_test
test_pred["pred_공종"] = y_pred
test_pred["pred_conf"] = y_conf

# 후보(Candiate), 유사도점수(Score) 라벨
test_pred["cand_공종_시료명vs종별"]  = cand_a[idx_test]
test_pred["cand_공종_종목vs시험종목"] = cand_b[idx_test]
test_pred["cand_공종_방법vs시험방법"] = cand_c[idx_test]
test_pred["score_시료명vs종별"]      = scr_a[idx_test]
test_pred["score_종목vs시험종목"]     = scr_b[idx_test]
test_pred["score_방법vs시험방법"]     = scr_c[idx_test]

display(test_pred.head(20))

# (선택) 저장
# os.makedirs("results", exist_ok=True)
# pd.DataFrame({"feature": feature_names, "importance": rf.feature_importances_}) \
#   .sort_values("importance", ascending=False) \
#   .to_csv("results/rf_feature_importance.csv", index=False, encoding="utf-8-sig")
test_pred.to_csv("results/test_predictions_with_all_columns.csv", index=False, encoding="utf-8-sig")

print("\n완료")
```
